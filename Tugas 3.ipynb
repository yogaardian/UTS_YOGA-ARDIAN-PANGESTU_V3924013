```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Membaca data dari Excel
file_path = "C:\\Users\\USER\\Downloads\\data_pasien2.csv"
df = pd.read_csv(file_path, encoding = 'windows-1252', delimiter=';')

# Tampilkan beberapa baris pertama untuk cek data
print(df.head())
```

```python
# a. Probabilitas benar-benar sakit jika hasil tes positif
positif = df[df['Hasil_Tes'] == 'Positif']

sakit_dan_positif = positif[positif['Menderita_Penyakit'] == 'Ya']
prob_sakit_dengan_tes_positif = len(sakit_dan_positif) / len(positif)

# Tampilkan hasil
print("\nHasil Perhitungan:")
print(f"Probabilitas pasien benar sakit jika hasil tesnya POSITIF: {prob_sakit_dengan_tes_positif:.2f}")
```

```python
# b. Probabilitas tidak sakit jika hasil tes negatif
negatif = df[df['Hasil_Tes'] == 'Negatif']
tidak_sakit_dan_negatif = negatif[negatif['Menderita_Penyakit'] == 'Tidak']
prob_tidak_sakit_dengan_tes_negatif = len(tidak_sakit_dan_negatif) / len(negatif)

# Tampilkan hasil
print("\nHasil Perhitungan:")
print(f"Probabilitas pasien tidak sakit jika hasil tesnya NEGATIF: {prob_tidak_sakit_dengan_tes_negatif:.2f}")
```

```python
# c. Visualisasi distribusi hasil tes dan status penyakit
plt.figure(figsize=(8, 5))

sns.countplot(data=df, x='Hasil_Tes', hue='Menderita_Penyakit', palette=['red', 'blue'])

plt.title("Distribusi Hasil Tes Positif dan Negatif serta Status Penyakit")
plt.xlabel("Hasil Tes")
plt.ylabel("Jumlah Pasien")
plt.legend(title="Menderita Penyakit")
plt.grid(True)
plt.show()
```
